## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)

df_ConvertedComp['ConvertedComp'].plot(kind='hist',color='g', figsize=(8, 5))
plt.title('Histogram of Converted Compensation')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual USD')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

QUERY = """
SELECT Respondent, Age
FROM master
"""
df_Age = pd.read_sql_query(QUERY,conn)

df_Age['Age'].describe()
```

```python
df_Age['Age'].plot(kind='box', figsize=(10, 7))

plt.title("Box plots of Respondent's Age")
plt.ylabel('Interval Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT * FROM master 
"""
df_corr = pd.read_sql_query(QUERY,conn)

plot = sns.scatterplot(x='Age', y='WorkWeekHrs', data=df_corr, color='g')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_bubble = pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=df_bubble, x='WorkWeekHrs', y='CodeRevHrs', hue='Age', alpha=0.5,color='g', sizes=(14, 8)) 

plt.title('Relation Between WorkWeekHrs and CodeRevHrs of Respondents', size=14) 
plt.xlabel('WorkWeekHrs', size=10) 
plt.ylabel('CodeRevHrs', size=10)
plt.legend(['Age'])

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

df_pie=pd.read_sql_query(QUERY,conn)  
df_pie.set_index('DatabaseDesireNextYear', inplace=True) 

df_pie['count'].plot(kind='pie', figsize=(5,6), autopct='%1.1f%%', startangle=90, shadow=False)  

plt.title('Top 5 Databases Respondents Wish To Learn Next Year') 
plt.axis('equal')

plt.show()
```
